```python
data = pd.read_csv('covid_19_data.csv') 
data = data.loc[data.Confirmed>=1]     
#data = data.loc[data.Country =='India']
df = data.dropna()   
    #prin(df.sort())  
death =0   
recover = 0  
cs =0 
#print(df)
for i in df.itertuples():
  cs += i.Confirmed
  death += i.Deaths 
  recover += i.Recovered
number = [cs, death , recover]  
labels = ["Confirmed Cases", "People Died" ,"Recovered Cases"] 
explode = (0.1 , 0.2 , 0.1)
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
slices = sorted(number)
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7)) 
theme = plt.get_cmap('copper') 
ax.set_prop_cycle("color", [theme(1. * i / len(number))
                             for i in range(len(number))])
wedges, texts, autotexts = ax.pie(number, 
                                  autopct = lambda pct: func(pct, number),
                                  explode = explode, 
                                  labels = labels,
                                  startangle = 90,
                                  textprops = dict(color ="w"))
 
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("World ratios of Confirmed, died and recovered people " , c = "w" , size = 15)  
# Adding legend
total = sum(number)
plt.legend(
    loc='lower left',
    labels=['%s -  %1.1f%%' % (
        l, (float(s) / total) * 100)
            for l, s in zip(labels, number)],
    prop={'size': 11},
    bbox_to_anchor=(0.0, 1),
    bbox_transform=fig.transFigure
)
  
  
# show plot
plt.show()
```
