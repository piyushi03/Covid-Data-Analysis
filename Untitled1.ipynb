```python
import numpy as np 
import numpy.random as rd 
import pandas as pd  
import matplotlib as mtp  
import matplotlib.pyplot as plt  
import sys 
import os
```

```python
data = pd.read_csv('covid_19_data.csv') 
data = data.loc[data.Confirmed>=1]    
data = data.loc[data.Country =='India']  
data.dropna(inplace = True) 
#print(data)  
states__ = data["State"] 
cases__ = data["Confirmed"]   
#print(cases__)
#print(states__) 
plt.pie(cases__ , labels = states__ , autopct='%2.2f%%', shadow=True, startangle=90) 
plt.title("Confirmed cases variation in states of India\n\n\n") 
plt.show()
```
